```python
import pyspark as ps
from pyspark.sql.types import BooleanType
from pyspark.sql.functions import udf
from pyspark.sql import functions as F
from pyspark.sql import SparkSession
from pyspark.sql import window
```

```python
alldata = spark.read.csv('data.csv',  header=True, inferSchema=True)
companies = spark.read.csv('comp.csv',  header=True, inferSchema=True)
```

```python
from pyspark.sql.functions import col
```

```python
data = alldata.select(col("description").alias("describe"),col('position').\
                      alias('position'),col('company').alias('company'),\
                      col('reviews').alias('reviews'),col('location').alias('location'))
data.show()
```

```python
comp = companies.join(data, companies['company name'] ==\
                      data.company)
```

```python
comp.select('ticker', 'company name', 'location', 'industry','describe').show()
```

```python
type(comp)
```

```python
comp.select('description').show()
```

```python
from pyspark.ml.feature import NGram,Tokenizer
```

```python
token = Tokenizer(inputCol ='describe', outputCol ='tokens')
comp1 =token.transform(comp)
```

```python
ngrams = NGram(n=2, inputCol ='tokens', outputCol ='ngrams')
compp = ngrams.transform(comp1)
```

```python
compp.columns
```

```python
compp.select('ngrams').show()
```

```python
compp.select(['location', 'ngrams']).show()
```
